```python
from bs4 import BeautifulSoup
from selenium import webdriver
from openpyxl import Workbook
```

```python
# 구글 코랩에서 크롬 드라이버 연동하기
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

driver = webdriver.Chrome('chromedriver', options = chrome_options)
```

```python
# 크롬 브라우저로 사이트 접속
page_url = 'https://www.billboard.com/charts/year-end/2020/hot-100-songs'
driver.get(page_url)
```

```python
# BeautifulSoup 모듈로 data select

# html = driver.page_source
# soup = BeautifulSoup(html, 'html.parser')

soup = BeautifulSoup(driver.page_source, 'html5lib')

songs = soup.select('div.ye-chart-item__title')
artists = soup.select('div.ye-chart-item__artist')

print(songs)
print(artists)
```

```python
song_list = []
artist_list = []

for song in songs:
    #print(song.text)
    song_list.append(song.text.strip('\n'))

print(song_list)

for artist in artists:
    artist_list.append(artist.text.strip('\n'))

print(artist_list)
```

```python
# Workbook 생성
wb = Workbook()
```

```python
# 시트 생성
ws1 = wb.active
ws1.title = "billboard"
```

```python
for i in range(0, len(song_list)):
    ws1['A' + str(i + 1)] = song_list[i]

for i in range(0, len(artist_list)):
    ws1['B' + str(i + 1)] = artist_list[i]
```

```python
# 데이터 저장
wb.save('billdata.tsv')
```

```python
# 크롬 드라이버 종료
driver.close()
```
